This is a complex format, 
 we need in this format -> ['Action', 'Adventure', 'Fantacy', 'Science Fiction']

```python
import ast
# to convert string to list
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)   # apply convert function on 'genres' column
```

```python
movies.head(2)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)   # apply convert function on 'keyword' column
```

```python
movies.head(2)
```

```python
movies['cast'][0]
```

```python
def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:     # will consider only first 3 casts
            L.append(i['name'])
            counter += 1
        else:
            break
                  
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)  # apply convert_cast function on 'cast' column
```

```python
movies.head(2)
```

```python
movies['crew'][0]
```

```python
# fetching director name from crew
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])         
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)  # apply fetch_director function on 'crew' column
```

```python
movies.head(2)
```

```python
movies['overview'][0]
```

```python
# convert overview strings to list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(3)
```

```python
# remove spaces 
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(2)
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] 
```

```python
movies.head(2)
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
# convert tags list to string
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
# convert tags to lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

#### text vectorization
bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')  # considered 5000 most frequent words, ignore english stopwords
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names()  # most frequent 5000 words
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

 for example - ps.stem('actions') = 'action'  ,  ps.stem('actioning') = 'action'

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

#### Calculating angle between movie vectors

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape 
```

Larger the angle smaller the similarity score between two movies and vice versa

```python
similarity[0]   # similarity score of 1st movie with each movie
```

```python
list(enumerate(similarity[0]))      # (movie_index, similarity_score_with 1st movie)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])  # sort in decreasing order of similarity score
```

```python
# recommend movie
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]  # index of required movie
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]   # Will consider only first 5 movies
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)  # names of recommended 5 movies
```

```python
recommend('Avatar')   # recommended movie names for movie 'Avatar'
```
